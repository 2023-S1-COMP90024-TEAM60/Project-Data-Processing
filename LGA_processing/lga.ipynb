```python
# Remember to import json
def process_lga_code(lga_path, twi_path):

    # eg. lga_path = "lga_list.json"
    # eg. twi_path = "sample_data_with_geo_loc.json"
    lga = json.load(open(lga_path,"r"))
    twi = json.load(open(twi_path,"r"))
    
    lga_dict = dict()
    state_dict = dict()
    new_twi = []

    for i in lga:
        name = i['lga_name_long'][0]
        code = i['lga_code'][0]
        state_code = str(i['ste_code'][0])
        state_name = i['ste_name'][0]
        lga_dict[name] = [code,state_code]
        if state_name not in state_dict:
            state_dict[state_name] = state_code
   
    for item in twi:
        name = item['doc']['includes']['places'][0]['full_name']
        name = name.split(', ')

        item['doc']['includes']['places'][0]['lga_name'] = '0'
        item['doc']['includes']['places'][0]['state_code'] = '0'
        item['doc']['includes']['places'][0]['lga_code'] = '0'

        if (name[0] in lga_dict):
            item['doc']['includes']['places'][0]['lga_code'] = lga_dict[name[0]][0]
            item['doc']['includes']['places'][0]['state_code'] = lga_dict[name[0]][1]
            item['doc']['includes']['places'][0]['lga_name'] = name[0]

        if (name[-1] in lga_dict):
            item['doc']['includes']['places'][0]['lga_code'] = lga_dict[name[-1]][0]
            item['doc']['includes']['places'][0]['state_code'] = lga_dict[name[-1]][1]
            item['doc']['includes']['places'][0]['lga_name'] = name[-1]

        if (name[0] in state_dict):
            item['doc']['includes']['places'][0]['state_code'] = state_dict[name[0]]
        if (name[-1] in state_dict):
            item['doc']['includes']['places'][0]['state_code'] = state_dict[name[-1]]  
    
        if not ((item['doc']['includes']['places'][0]['state_code'] == '0') and (item['doc']['includes']['places'][0]['state_code'] == '0')):
            new_twi.append(item)

    return new_twi
    
    
```
